ATIVIDADE:

Construa um jogo da velha para dois jogadores que possua as seguintes características:

1 - O jogo não será multiplayer. Logo um jogador jogará contra a maquina. \\
2 - A função de decisão para a maquina deverá implementar um agente reativo simples. Logo, as decisões das jogadas não devem considerar seu histórico de açoes anteriores para decidir qual a melhor jogada a ser realizada.  **Sugestão: A jogada pode ser aleatoria dentro dos campos ainda liberados para serem escolhidos.** \\
3 - Caberá ao aluno definir quem irá realizar a primeira ação (Jogador ou Máquina). Após isso as jogadas serão alternadas entre Jogador e Maquina.

4 - Não é obrigatório  implementar interface gráfica.

```python
import random

tab = [0, 1, 2, 3, 4, 5, 6, 7, 8]

def desenho():
    print(tab[0], '|', tab[1], '|', tab[2])
    print('-' * 9)
    print(tab[3], '|', tab[4], '|', tab[5])
    print('-' * 9)
    print(tab[6], '|', tab[7], '|', tab[8])

def humano():
    while True:
        escolha = int(input('Escolha a posicao: '))
        break
    tab[escolha] = 'O'


def jogadaIA():
    print('Jogada da IA:')
    opcoesValidas = []
    for i in range(0, 8):
        if tab[i] != 'X' and tab[i] != 'O':
            opcoesValidas.append(i)
    opcao = random.choice(opcoesValidas)
    tab[opcao] = 'X'


def ganhou(turno):
      if(
          (tab[0] == turno and tab[1] == turno and tab[2] == turno) or
          (tab[3] == turno and tab[4] == turno and tab[5] == turno) or
          (tab[6] == turno and tab[7] == turno and tab[8] == turno) or
          (tab[0] == turno and tab[3] == turno and tab[6] == turno) or
          (tab[1] == turno and tab[4] == turno and tab[7] == turno) or
          (tab[2] == turno and tab[5] == turno and tab[8] == turno) or
          (tab[0] == turno and tab[4] == turno and tab[8] == turno) or
          (tab[2] == turno and tab[4] == turno and tab[6] == turno)
      ):
        return True
      else:
        return False

desenho()

turnoHumano = True

for i in range(0, 9):
    if turnoHumano:
        humano()
        if ganhou('O'):
            desenho()
            print('Jogador Humano ganhou!')
            break
    else:
        jogadaIA()
        if ganhou('X'):
            desenho()
            print('IA ganhou!')
            break
    desenho()
    turnoHumano = not turnoHumano

    if i == 8:
        print('Deu velha')
        print('Fim do jogo.')
        break
```
